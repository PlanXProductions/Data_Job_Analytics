Roles to explore:

```python
# Importing Libraries
import ast
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt 
from adjustText import adjust_text
import seaborn as sns


# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
df['job_skills'] = df['job_skills'].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else x) 
```

Filter for US Data Analyst roles:

```python
df_DA_US = df[(df['job_title_short'] == 'Data Analyst') & (df['job_country'] == 'United States')]
df_DA_US
```

Columns to Analyze:
- job_location (top 10 within US)

```python
df_DA_US['job_location'].value_counts().head(10)
```

Plot

```python
# create df from filter:
df_plot = df_DA_US['job_location'].value_counts().head(10).to_frame()

sns.barplot(data=df_plot, x='count', y='job_location')
```

```python
# create df from filter:
df_plot = df_DA_US['job_location'].value_counts().head(10).to_frame()

# color and label
sns.set_theme(style='ticks')
sns.barplot(data=df_plot, x='count', y='job_location', hue='count', palette='dark:b_r', legend=False)
# removes border box
sns.despine()
plt.title('Counts of Job Locations for US-based Data Analyst')
plt.xlabel('Number of Jobs')
plt.ylabel('')
plt.show()
```

Explore WFH, degree, health insurance:
- ref. 2_14

```python
# rewrite the above with a for loop
dict_column = {
    'job_work_from_home': 'Work from Home Offered',
    'job_no_degree_mention': 'Degree Requirement',
    'job_health_insurance': 'Health Insurance Offered'
}

fig, ax = plt.subplots(1, 3)
fig.set_size_inches((12, 5))

for i, (column, title) in enumerate(dict_column.items()):
    ax[i].pie(df_DA_US[column].value_counts(), labels=['False', 'True'], autopct='%1.1f%%', startangle=90)
    ax[i].set_title(title)

# plt.suptitle('Benefit Analysis of Data Jobs', fontsize=16)
plt.show()
```

Company names:

```python
# create df from filter:
df_plot = df_DA_US['company_name'].value_counts().head(10).to_frame()

# color and label
sns.set_theme(style='ticks')
sns.barplot(data=df_plot, x='count', y='company_name', hue='count', palette='dark:b_r', legend=False)
# removes border box
sns.despine()
plt.title('Companies with roles for US-based Data Analyst')
plt.xlabel('Number of Jobs')
plt.ylabel('')
plt.show()
```
